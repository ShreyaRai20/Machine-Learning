```python
import numpy as np
```

```python
arr = np.array([2,3,4,6,7,8,9,12, 13, 16, 17,23, 25,27,34, 37,205])
arr
```

```python
Q1 = np.percentile(arr,25)
Q3 = np.percentile(arr,75)
IQR = Q3 - Q1

print(f'Q1: {Q1}, Q3: {Q3}, IQR: {IQR}')
```

```python
UF = Q3 + (1.5 * IQR)
LF = Q1 - (1.5 * IQR)
print(f'UF: {UF}, LF: {LF}')
```

```python
l = []
for i in arr:
    if( i >= LF and i <= UF):
        l.append(i)

arr2 = np.array(l)
arr2
```
