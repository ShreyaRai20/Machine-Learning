```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

warnings.filterwarnings('ignore')
```

```python
X, y_true = make_blobs(n_samples=500,centers=3 ,cluster_std=0.50, random_state=42)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
df = pd.DataFrame(X_scaled)
df
```

```python
inertia = []
k_range = range(1,11)
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
sns.lineplot(x=k_range, y=inertia, markers='x')
```

```python
kmeans_final = kmeans = KMeans(n_clusters=3,random_state=42)
```

```python
cluster_labels = kmeans_final.fit_predict(X_scaled)
```

```python
cluster_labels
```

```python
df['cluser'] = cluster_labels
```

```python
df
```

```python
sns.scatterplot(data=df, x=df[0], y=df[1], hue=df['cluser'], palette='viridis')
```
